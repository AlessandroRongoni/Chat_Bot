```python
import pandas as pd

# Carica il dataset
df = pd.read_csv('../dataset/cleaned/netflix_titles_cleaned.csv')

# Assumiamo che la colonna che contiene i paesi si chiami 'country'
# Splitta i paesi basandoti sul delimitatore (es. una virgola) e poi espandi la lista
all_countries = df['country'].str.split(', ', expand=True)

# Raccogli tutti i paesi unici trasformando il DataFrame in un long format e rimuovendo i duplicati
unique_countries = pd.unique(all_countries.values.ravel('K'))

# Rimuovi i valori NaN (se presenti)
unique_countries = [country for country in unique_countries if pd.notna(country)]

# Stampa la lista dei paesi unici
for country in sorted(unique_countries):
    print(country)
```
